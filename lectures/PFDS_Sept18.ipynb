```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read csv file :
df = pd.read_csv("http://rcs.bu.edu/examples/python/DataAnalysis/Salaries.csv")
```

```python
df.info()
```

```python
df.info
```

```python
df.rename(columns = {'rank':'designation'}, inplace = True)
```

```python
df.info()
```

```python
#Use matplotlib to draw a histogram of a salary data
plt.hist(df['salary'],bins=8, density=True);
```

```python
#Use seaborn package to draw a histogram
sns.distplot(df['salary']);
```

```python
sns.histplot(df['salary']);
```

```python
# Matplotlib function to display a barplot :
df.groupby(['service'])['salary'].count().plot(kind='bar')
```

```python
df.head(10)
```

```python
df.groupby(['designation'])['salary'].count().plot(kind='bar')
```

```python
# Barplot using seaborn :
sns.set_style("whitegrid")

ax = sns.barplot(x='designation',y ='salary', data=df, estimator=np.mean,ci=None)
```

```python
# Split into two groups :
ax = sns.barplot(x='designation',y ='salary', hue='sex', data=df)#, estimator=len)
```

```python
# Pairplot 
sns.pairplot(df)
```

```python
#Scatterplot using seaborn :
sns.jointplot(x='service', y='salary', data=df)
```

```python
#Linear regression plot for two numeric variables using regplot :
sns.regplot(x='service', y='salary', data=df)
```

```python
# box plot :
sns.boxplot(x='designation',y='salary', data=df)
```

```python
# swarm plot :
sns.swarmplot(x='designation',y='salary', data=df)
```
