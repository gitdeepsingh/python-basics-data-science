```python
# ANSWER 1

grade = 'A' # as maximum marks are not mentioned in the question, default grade is A
try:
    inputMarks = int(input('Please enter marks: '))
    if(inputMarks < 25):
      grade = 'F'
    if(25 <= inputMarks < 45):
      grade = 'E'
    elif(45<= inputMarks < 50):
      grade = 'D'
    elif(50 <= inputMarks < 60):
      grade = 'C'
    elif(60 <= inputMarks < 80):
      grade = 'B'
    print(f'Yoour Grade is {grade}')
except ValueError: # catch non-numeric inputs appropriately
    print('Invalid marks input. Please re-try!')
```

```python
# Answer 4

def is_prime(n):
    isPrime = True
    if n <=1:
        isPrime = False
    for i in range(2, int(n)):
        if (n % i == 0):
             isPrime = False
    return isPrime


# example
is_prime(2)
```

```python
#Answer 5

def get_frequency_dic(itemList):
    freq_dict = {}
    for item in itemList:
        if(item in freq_dict):
            freq_dict[item] += 1
        else:
            freq_dict[item] = 1
    return freq_dict
        
# example
get_frequency_dic(["here", "there",2, "where", "here", "here", "there", 2,99,2])
```

```python
# Answer 6

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (np.sin(x-2)**2) * (np.exp(-x ** 2))
    
x = np.linspace(0, 2, 1000)
y=f(x)

plt.plot(x, y, label='f(x)')
plt.title('f(x)=$sin^{2}(x-2)$ $e^{-x^2}$')
plt.grid(c='lightgray')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
```
